```python
from sklearn import datasets
from matplotlib import pyplot
from pandas import DataFrame
# 生成2d分类数据集
X, Y = datasets.make_moons(n_samples=100, noise=0.1)
# 散点图，按分类值着色
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=Y))
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
  print(key)
  group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
 
pyplot.show()
```

```python
train_count = len(X)

np.random.seed(1)

w = np.random.randn(3,1)

epoch = 1000
etha = 0.05

for i in range(epoch):
  cost = 0
  for j in range(train_count):
    x1 = X[j][0]
    x2 = X[j][1]
    y = Y[j]
    output = 1*w[0] + x1*w[1] +x2*w[2]
    cost +=(y-output)*(y-output)
   
    deltaw1 = (y-output) *1
    deltaw2 = (y-output) *x1
    deltaw3 = (y-output) *x2
    w[0] += etha*(deltaw1/train_count)
    w[1] += etha*(deltaw2/train_count)
    w[2] += etha*(deltaw3/train_count)
  cost = cost/train_count
  print("epoch: "+str(i)+"cost: "+str(cost)+"W: "+str(w))

print("training done")
```

```python
t = np.arange(-1, 2, 0.01)
s = -(1 * w[0] + t*w[1])/w[2]

df = DataFrame(dict(x=X[:,0], y=X[:,1], label=Y))
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')

for key, group in grouped:
  group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
 
pyplot.plot(t, s, color='black', ls='-')
pyplot.show()
```
